## 『本次練習目的』
  #### 本次練習主要是要讓各位同學們了解CNN與FC層的參數使用量差異
  #### Convolution2D有許多參數可以設置，之後章節會細談
  #### 不熟Keras的學員們也可以藉此了解NN層的不同搭法

```python
from keras.models import Sequential
from keras.layers import Convolution2D
from keras.layers import Input, Dense
from keras.models import Model
```

```python
##輸入照片尺寸=28*28*1
##都用一層，288個神經元

##建造一個一層的CNN層
classifier = Sequential()

##Kernel size 3*3，用32張，輸入大小28*28*1
'''32張Kernel，大小為3*3，輸入尺寸為28*28*1'''
classifier.add(Convolution2D(32, (3, 3), input_shape=(28, 28, 1)))

##看看model結構
print(classifier.summary())

'''理解輸出Total params為何=320'''
# Total params = (Kernel_H*W*channels+1)*Kernel numbers
Kernel_H = 3
Kernel_W = 3
channels = 1
Kernel_numbers = 32
print("Total params:", (Kernel_H*Kernel_W*channels+1)*Kernel_numbers)

##建造一個一層的FC層
classifier = Sequential()

##輸入為28*28*1攤平==784
'''輸入尺寸為28*28*1'''
inputs = Input(shape=(28*28*1,))

##CNN中用了(3*3*1)*32個神經元，我們這邊也用相同神經元數量
'''使用288個神經元'''
x = Dense(288)(inputs)

model = Model(inputs=inputs, outputs=x)

##看看model結構
print(model.summary())

'''理解輸出Total params為何=226080'''
# Total params = Input*output+output (bias)
_Input = 784
_output = 288
print("Total params:", _Input*_output+_output)

##大家可以自己變換設定看看參數變化
```
